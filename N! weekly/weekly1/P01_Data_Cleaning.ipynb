7. Визуализируйте данные с предыдущего задания линейным графиком, где должны быть соблюдены следующие условия:
    - Период январь-апрель должен быть зеленого цвета
    - Период май-август должен быть красного цвета
    - на линейном графике должны присутствовать только средние значения (mean)

```python
from datetime import datetime

low1 = datetime(2021,1,1)
high1 = datetime(2021,4,30)

low2 = datetime(2021,5,1)
high2 = datetime(2021,9,30)

first_period = df[(pd.to_datetime(df['timestamp']) >= low1) & ((pd.to_datetime(df['timestamp']) <= high1)) & (df['title'] == 'Software Engineer')]
```

```python
print(f"mean: {first_period['totalyearlycompensation'].mean()}")
print(f"median: {first_period['totalyearlycompensation'].median()}")
print(f"mode: {first_period['totalyearlycompensation'].mode()}")
```

```python
second_period = df[(pd.to_datetime(df['timestamp']) >= low2) & ((pd.to_datetime(df['timestamp']) <= high2)) & (df['title'] == 'Software Engineer')]
print(f"mean: {second_period['totalyearlycompensation'].mean()}")
print(f"median: {second_period['totalyearlycompensation'].median()}")
print(f"mode: {second_period['totalyearlycompensation'].mode()}")
```

```python
import matplotlib.pyplot as plt

first_q = pd.to_datetime(first_period['timestamp']).dt.month.unique()
second_q = pd.to_datetime(second_period['timestamp']).dt.month.unique()

first_means = []
for month in first_q:
    x = first_period[pd.to_datetime(first_period['timestamp']).dt.month == month]['totalyearlycompensation'].mean()
    first_means.append(x)
    
sec_means = []
for month in second_q:
    x = second_period[pd.to_datetime(second_period['timestamp']).dt.month == month]['totalyearlycompensation'].mean()
    sec_means.append(x)
    
plt.plot(first_means, c='green')
plt.plot(sec_means, c='red')

plt.xlabel('by month in quater')
plt.ylabel('wage mean')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

#### 10. Сделайте сравнение уровня зарплат в зависимости от грэйда (level) в компании Amazon на позицию 'Software Engineer'. Постройте визуализацию и напишите ваши выводы.

```python
amazon = df[(df['company'] == 'Amazon') & (df['level'].notnull()) & (df['title'] == 'Software Engineer')]
```

```python
grades = amazon['level'].unique()
```

```python
for grade in grades:
    all_data = amazon[amazon['level'] == grade]['totalyearlycompensation']
    print(f"stats by grade - {grade}: mean - {all_data.mean()}, max - {all_data.max()}, min - {all_data.min()}, count - {all_data.count()}")
```

```python
import matplotlib.pyplot as plt

levels = ['L4', 'L5', 'L6', 'SDE I', 'SDE II', 'SDE III']

for grade in grades:
    all_data = amazon[amazon['level'] == grade]['totalyearlycompensation']
    y = [levels.index(grade) for w in all_data]
    x = [int(w) for w in all_data]
    plt.scatter(x, y)
plt.show()
print([f"{i} - {levels[i]}" for i in range(len(levels))])
```

As it is seen in average the higher the grading the higher is wage, but still there is a lot of cases when for example you are L4 engineer and you earn more than L6 experts. So you can not be sure if you earn more than other just by grading level
