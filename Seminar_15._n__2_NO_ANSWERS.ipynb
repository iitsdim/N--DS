# Метрики

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=200, n_classes=3, n_features=2, n_informative=2,
                           n_redundant=0, n_clusters_per_class=1, weights=[0.6,0.3,0.1],random_state=42)

# Визуализация данных
plt.figure(figsize=(14, 8))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.title('Задача классификации - пример данных')
plt.show()

pd.Series(y).value_counts()
```

```python
# Функция для визуализации границы принятия решений
def plot_decision_boundary(model, title):
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
from sklearn.svm import SVC
```

```python
# Создание и обучение модели
model_log_multi = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model_log_multi.fit(X, y)

model_log_ovr = LogisticRegression(multi_class='ovr', solver='liblinear')
model_log_ovr.fit(X, y)

model_ovr_log = OneVsRestClassifier(LogisticRegression(solver='liblinear'))
model_ovr_log.fit(X, y)

model_ovo_log = OneVsOneClassifier(LogisticRegression(solver='liblinear'))
model_ovo_log.fit(X, y)
```

```python
plt.figure(figsize=(15, 10))

# Визуализация границы принятия решений
plt.subplot(2, 2, 1)
plot_decision_boundary(model_log_multi, title="Multinomial Logistic Regression")

plt.subplot(2, 2, 2)
plot_decision_boundary(model_log_ovr, title="Logistic Regression One-vs-Rest")

plt.subplot(2, 2, 3)
plot_decision_boundary(model_ovr_log, title="Binary Logistic Regression (One-vs-Rest)")

plt.subplot(2, 2, 4)
plot_decision_boundary(model_ovo_log, title="Binary Logistic Regression (One-vs-One)")
```

```python
plt.figure(figsize=(15, 10))

# Визуализация границы принятия решений
plt.subplot(2, 2, 1)
plot_decision_boundary(model_log_multi, title="Multinomial SVC")

plt.subplot(2, 2, 2)
plot_decision_boundary(model_log_ovr, title="SVC One-vs-Rest")

plt.subplot(2, 2, 3)
plot_decision_boundary(model_ovr_log, title="Binary SVC (One-vs-Rest)")

plt.subplot(2, 2, 4)
plot_decision_boundary(model_ovo_log, title="Binary SVC (One-vs-One)")
```

## Accuracy. Classification report

```python
# Вычисление матрицы ошибок
cm = confusion_matrix(y, y_pred)
class_names = ['Class 0', 'Class 1', 'Class 2']
tick_marks = np.arange(len(class_names))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Вычисление метрик качества
accuracy = accuracy_score(y, y_pred)
balanced_accuracy = balanced_accuracy_score(y, y_pred)
classification_rep = classification_report(y, y_pred)

# Вывод результатов
print("Accuracy: %0.2f" % accuracy)
print("Balanced Accuracy: %0.2f" % balanced_accuracy)
print("Classification Report:\n", classification_rep)
```

## Precision

```python
precision_micro = precision_score(y, y_pred, average='micro')
precision_macro = precision_score(y, y_pred, average='macro')
precision_weighted = precision_score(y, y_pred, average='weighted')

print("Precision Micro: %0.2f" % precision_micro)
print("Precision Macro: %0.2f" % precision_macro)
print("Precision Weighted: %0.2f" % precision_weighted)
```

## Recall

```python
recall_micro = recall_score(y, y_pred, average='micro')
recall_macro = recall_score(y, y_pred, average='macro')
recall_weighted = recall_score(y, y_pred, average='weighted')

print("Recall Micro: %0.2f" % recall_micro)
print("Recall Macro: %0.2f" % recall_macro)
print("Recall Weighted: %0.2f" % recall_weighted)
```

## F1 score

F1 = 2 * (precision * recall) / (precision + recall)

```python
f1_micro = f1_score(y, y_pred, average='micro')
f1_macro = f1_score(y, y_pred, average='macro')
f1_weighted = f1_score(y, y_pred, average='weighted')

print("F1 Micro: %0.2f" % f1_micro)
print("F1 Macro: %0.2f" % f1_macro)
print("F1 Weighted: %0.2f" % f1_weighted)
```

``'binary'``:<br>
    Only report results for the class specified by ``pos_label``.
    This is applicable only if targets (``y_{true,pred}``) are binary.<br>
``'micro'``:<br>
    Calculate metrics globally by counting the total true positives
    false negatives and false positives.<br>
``'macro'``:<br>
    Calculate metrics for each label, and find their unweighted
    mean.  This does not take label imbalance into account.<br>
``'weighted'``:<br>
    Calculate metrics for each label, and find their average weighted
    by support (the number of true instances for each label). This
    alters 'macro' to account for label imbalance; it can result in an
    F-score that is not between precision and recall.<br>
``'samples'``:<br>
    Calculate metrics for each instance, and find their average (only
    meaningful for multilabel classification where this differs from
    :func:`accuracy_score`).<br>

## Пример вычисления метрик

```python
# Предсказанные и истинные метки классов
y_true = [0, 1, 2, 0, 1, 2]
y_pred = [0, 2, 1, 0, 0, 1]

# Вычисление TP, FP и support для каждого класса
tp = [0] * len(set(y_true))
fp = [0] * len(set(y_true))
fn = [0] * len(set(y_true))
support = [0] * len(set(y_true))

for true, pred in zip(y_true, y_pred):
    if true == pred:
        tp[true] += 1
    else:
        fp[pred] += 1
        fn[true] += 1        
    support[true] += 1

# Вычисление метрик precision micro, precision macro и precision weighted
tp_total = sum(tp)
fp_total = sum(fp)
fn_total = sum(fn)
```

```python
precision_micro = tp_total / (tp_total + fp_total)
precision_macro = sum(tp[i] / (tp[i] + fp[i]) for i in range(len(tp))) / len(tp)
precision_weighted = sum(tp[i] / (tp[i] + fp[i]) * support[i] for i in range(len(tp))) / sum(support)

# Вывод результатов
print("Precision micro:", precision_micro)
print("Precision macro:", precision_macro)
print("Precision weighted:", precision_weighted)
```

```python
recall_micro = tp_total / (tp_total + fn_total)
recall_macro = sum(tp[i] / (tp[i] + fn[i]) for i in range(len(tp))) / len(tp)
recall_weighted = sum(tp[i] / (tp[i] + fn[i]) * support[i] for i in range(len(tp))) / sum(support)

# Вывод результатов
print("Recall micro:", recall_micro)
print("Recall macro:", recall_macro)
print("Recall weighted:", recall_weighted)
```

```python
f1_micro = 2 * (precision_micro * recall_micro) / (precision_micro + recall_micro)
f1_macro = 2 * (precision_macro * recall_macro) / (precision_macro + recall_macro)
f1_weighted = sum(tp[i] / (tp[i] + fp[i]) * support[i] for i in range(len(tp))) / sum(support)

# Вывод результатов
print("F1 micro:", f1_micro)
print("F1 macro:", f1_macro)
print("F1 weighted:", f1_weighted)
```

```python
# Вычисление метрик F1 micro, F1 macro и F1 weighted
f1_micro = f1_score(y_true, y_pred, average='micro')
f1_macro = f1_score(y_true, y_pred, average='macro')
f1_weighted = f1_score(y_true, y_pred, average='weighted')

# Вывод результатов
print("F1 micro:", f1_micro)
print("F1 macro:", f1_macro)
print("F1 weighted:", f1_weighted)
```
