```python
from sklearn.datasets import load_digits #MNIST
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Загрузка данных
digits = load_digits()
```

```python
digits.data[0]
```

```python
%%time
# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, 
                                                    test_size=0.3, random_state=1437)
```

```python
# Обучение модели
clf = SVC() # экземляр объекта Support Vector Machines
clf.fit(X_train, y_train)
```

```python
# Предсказание на тестовой выборке
y_pred = clf.predict(X_test)
```

```python
# Оценка точности
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
%%time
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# генерируем случайные данные
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=1437)

# создаем объект KMeans и обучаем модель
kmeans = KMeans(n_clusters=4) # попробуйте изменить кол-во допустимых кластеров
kmeans.fit(X)

# визуализируем кластеры
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, 
                                            c='black', label='Centroids')
plt.legend()
plt.show()
```

```python
for i in range(1, 10):
    # создаем объект KMeans и обучаем модель
    kmeans = KMeans(n_clusters=i) # попробуйте изменить кол-во допустимых кластеров
    kmeans.fit(X)

    # визуализируем кластеры
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, 
                                                c='black', label='Centroids')
    plt.legend()
    plt.show()
```
