## Комбинирование конструкций

#### 1. Посчитать количество аварий, произошедших в каждом городе и вывести список городов в порядке убывания количества аварий.

```python
sql_query = """
SELECT city, COUNT(*) AS accidents_count
FROM contracts
WHERE accident_date IS NOT NULL
GROUP BY city
ORDER BY accidents_count DESC
"""
execute_sql(sql_query)
```

```python
result = contracts \
    .loc[contracts['accident_date'].notnull(), ['city', 'accident_date']]\
    .groupby('city')\
    .count()\
    .sort_values('accident_date', ascending=False)\
    .rename(columns={'accident_date': 'accidents_count'}) \
    .reset_index()
result
```

#### 2. Вычислить количество записей для каждой марки автомобиля, выпущенной в 2010 году или позже, и вывести список марок в порядке убывания количества записей (только топ 10 марок)

```python
sql_query = """
SELECT brand, COUNT(*) AS count 
FROM contracts 
WHERE year_of_manufacture >= 2010 
GROUP BY brand 
ORDER BY count(*) desc
LIMIT 10
"""
execute_sql(sql_query)
```

```python
result = contracts \
    .loc[contracts['year_of_manufacture'].dt.year >= 2010, ['brand', 'year_of_manufacture']] \
    .groupby('brand')\
    .count()\
    .sort_values('year_of_manufacture', ascending=False)\
    .rename(columns={'year_of_manufacture': 'count'})\
    .head(10) \
    .reset_index()
result
```

### Batch 2.

#### 1. Вывести список уникальных комбинаций марок и моделей автомобилей, встречающихся в датасете, а также количество записей для каждой комбинации в порядке убывания количества записей (вывести только топ 10). Формат вывода: brand - model - count

```python
sql_query = """
SELECT brand, model, COUNT(*) AS count
FROM contracts
GROUP BY brand, model
ORDER BY count DESC
LIMIT 10
"""
execute_sql(sql_query)
```

#### 2. Вычислить средний возраст водителей для каждого города, в котором произошли аварии, и вывести список городов в порядке убывания среднего возраста.

```python
sql_query = """
SELECT city, AVG(age) AS avg_age
FROM contracts 
WHERE accident_date IS NOT NULL
GROUP BY city
ORDER BY AVG(age) DESC
"""
execute_sql(sql_query)
```

#### 3 Вывести список уникальных типов транспортных средств, участвовавших в авариях, а также количество записей для каждого типа в порядке убывания количества записей.

```python
sql_query = """
SELECT vehicle_type, COUNT(*) AS count
FROM contracts
WHERE accident_date IS NOT NULL
GROUP BY vehicle_type
ORDER BY COUNT(*) DESC
"""
execute_sql(sql_query)
```

#### 4. Вычислить среднюю сумму убытков для каждого города, в котором произошли аварии, и вывести список городов в порядке убывания средней суммы.

```python
sql_query = """
SELECT city, AVG(loss_amount) AS avg_loss_amount
FROM contracts
WHERE loss_amount > 0
GROUP BY city
ORDER BY AVG(loss_amount) DESC
"""
execute_sql(sql_query)
```
